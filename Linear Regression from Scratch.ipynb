```python
df = pd.read_csv('dataset.csv')

print(df.shape)
df.head()
```

```python
X= df['Head Size(cm^3)'].values
Y=df['Brain Weight(grams)'].values

X_mean=np.mean(X)
Y_mean= np.mean(Y)
n= len(X)
```

```python
plt.scatter(X,Y, color= 'cyan')
plt.title('Head Size vs Brain Weight')
plt.xlabel("Head Size")
plt.ylabel("Brain Weight")
plt.show()
```

```python
numerator=0
denominator=0
for i in range(n):
    numerator+=(X[i]-X_mean)*(Y[i]-Y_mean)
    denominator+= (X[i]-X_mean)**2

m= numerator/denominator
c= Y_mean- m*X_mean

m,c
```

```python
a= np.array([[1,2,3,4,5]])
b= np.array([[2,4,5,4,5]])

a_mean=np.mean(a)
b_mean=np.mean(b)

numerator=0
denominator=0
for i in range(n):
    numerator+=(a[i]-a_mean)*(b[i]-Y_mean)
    denominator+= (a[i]-a_mean)**2

m= numerator/denominator
c= Y_mean- m*X_mean

m,c
```

```python
#plotting values 
x_max = np.max(X) + 100
x_min = np.min(X) - 100
#calculating line values of x and y

y_pred = c + m * X

#plotting line 
plt.plot([x_min,x_max], [min(y_pred), max(y_pred)], color='#00ff00', label='Linear Regression')


#plot the data point
plt.scatter(X, Y, color='#ff0000', label='Data Point')
# x-axis label
plt.xlabel('Head Size (cm^3)')
#y-axis label
plt.ylabel('Brain Weight (grams)')
plt.legend()
plt.show()
```

```python
sumofsquares = 0
sumofresiduals = 0
for i in range(n) :
    y_pred = c + m * X[i]
    sumofsquares += (Y[i] - Y_mean) ** 2
    sumofresiduals += (Y[i] - y_pred) **2
    
score  =  (sumofresiduals/sumofsquares)
print(score)
```
