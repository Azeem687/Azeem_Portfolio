# Line Graph

```python
gas= pd.read_csv('gas_prices.csv')
plt.figure(figsize= (8,5))
plt.title('Gas Prices over Time (in USD)')


# plt.plot(gas.Year, gas.USA, 'b.-', label = "US") 

# plt.plot(gas.Year, gas.South Korea, 'r.-', label= 'South Korea')
# plt.plot(gas.Year, gas.Canada, 'g.-', label= 'Canada')

for country in gas:
    if country != "Year":
        plt.plot(gas.Year, gas[country], marker= '.', label= country)
    
# lets make x ticks for every third year
plt.xticks(gas.Year[::3])
# x and y label
plt.xlabel('Year')
plt.ylabel('US $')



plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```

# ###PIE Chart

```python
labels = 'Left', 'Right'
colors= ['#42f4df', 'red']
left= fifa[fifa['Preferred Foot']== 'Left'].count()[0]
right= fifa[fifa['Preferred Foot']== 'Right'].count()[0]
left_or_right= [left, right]

plt.figure(figsize=(8,5), dpi=100)
plt.style.use('ggplot')

plt.pie(left_or_right, labels= labels, autopct='%1.1f%%',
        shadow=True, colors= colors, startangle=90)

plt. axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

# MORE PIE CHART

```python


plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')



light = fifa[fifa.Weight < '125'].count()[0]
light_medium = fifa[(fifa.Weight >= '125') & (fifa.Weight < '150')].count()[0]
medium = fifa[(fifa.Weight >= '150') & (fifa.Weight < '175')].count()[0]
medium_heavy = fifa[(fifa.Weight >= '175') & (fifa.Weight < '200')].count()[0]
heavy = fifa[fifa.Weight >= '200'].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

# Scoring Distribution based on Draft Round

```python
nba= pd.read_csv('all_seasons.csv')

plt.figure(figsize=(8,5), dpi=100)
one = nba[(nba.draft_round=='1')& (nba.pts < 3)].count()[0]
two = nba[(nba.draft_round=='1')& (nba.pts > 3)& (nba.pts < 10)] .count()[0]
three = nba[(nba.draft_round=='1')& (nba.pts > 10) & (nba.pts < 15)] .count()[0]
four = nba[(nba.draft_round=='1')& (nba.pts > 15) & (nba.pts < 20)] .count()[0]
five= nba[(nba.draft_round=='1')& (nba.pts > 20) & (nba.pts < 100)] .count()[0]

Scoring = [one, two, three, four, five]
label = ['under 3 points', '3-10', '10-15', '15-20', '20+']


plt.title('Scoring Distribution of 1st round picks')

plt.pie(Scoring, labels=label, pctdistance=0.8,autopct='%.2f %%')
plt.show()
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')




badScorers = nba[(nba.draft_round=='2')& (nba.pts < 3)].count()[0]
alright = nba[(nba.draft_round=='2')& (nba.pts > 3)& (nba.pts < 10)] .count()[0]
prettySolid = nba[(nba.draft_round=='2')& (nba.pts > 10) & (nba.pts < 15)] .count()[0]
prettyGood = nba[(nba.draft_round=='2')& (nba.pts > 15) & (nba.pts < 20)] .count()[0]
Amazing= nba[(nba.draft_round=='2')& (nba.pts > 20) & (nba.pts < 100)] .count()[0]

Scoring = [badScorers, alright, prettySolid, prettyGood, Amazing]
label = ['under 3 points', '3-10', '10-15', '15-20', '20+']


plt.title('Scoring Distribution of 2nd round picks')

plt.pie(Scoring, labels=label, pctdistance=0.8,autopct='%.2f %%')
plt.show()




```

```python
plt.figure(figsize=(8,5), dpi=100)
one1 = nba[(nba.draft_round=='Undrafted')& (nba.pts < 3)].count()[0]
two2= nba[(nba.draft_round=='Undrafted')& (nba.pts > 3)& (nba.pts < 10)] .count()[0]
three3 = nba[(nba.draft_round=='Undrafted')& (nba.pts > 10) & (nba.pts < 15)] .count()[0]
four4 = nba[(nba.draft_round=='Undrafted')& (nba.pts > 15) & (nba.pts < 20)] .count()[0]
five5= nba[(nba.draft_round=='Undrafted')& (nba.pts > 20) & (nba.pts < 100)] .count()[0]

Scoring = [one1, two2, three3, four4]
label = ['under 3 points', '3-10', '10-15', '15-20']


plt.title('Scoring Distribution of undrafted players')

plt.pie(Scoring, labels=label, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```
