```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score 
x= np.array([1,2,3,4,5,6,7,8])
y=np.array([2,4,8,16,32,64,128,256])
linreg= LinearRegression()
x=x.reshape(-1,1)

linreg.fit(x,y)
y_pred= linreg.predict(x)


```

```python
plt.scatter(x,y, color= 'red')
plt.plot(x,y_pred, color= 'cyan')
plt.yticks([0,20,40,60,80,100,120,140,160,180,200,220,240])

plt.show()
```

```python
r2= r2_score(y, y_pred)
print(r2)
```

```python
print("The slope is", linreg.coef_, "and the intercept is",  linreg.intercept_)
```

# Linear Regression with Multiple Variables

```python
import math
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

df= pd.read_csv("hiring.csv")

df
```

```python
df= df.dropna(how='all', axis='columns')
df= df.dropna(how='all', axis='rows')
df

```

```python
median_testscore= math.floor(df.test_score.median())
median_testscore
```

```python
df.test_score_outof10= df.test_score.fillna(median_testscore)
df
```

```python
df['experience']= df['experience'].fillna(0)
df['experience']= df['experience'].astype(str)
df
```

```python
df.experience= df.experience.replace(["five"], 5)
df.experience= df.experience.replace(["two"], 2)
df.experience= df.experience.replace(["seven"], 7)
df.experience= df.experience.replace(["three"], 3)

df.experience= df.experience.replace(["eleven"], 11)
df.drop(6, inplace= True)
df
```

```python
df.loc[8]=[10,median_testscore,7,72000]
df
```

```python

df['experience']= df['experience'].astype(int)
df['salary']= df['salary'].astype(int)
df['test_score']= df['test_score'].astype(int)
df['interview_score']= df['interview_score'].astype(int)
df
```

```python
## shape all the column's data
experience= df.experience.values.reshape(-1,1)
test_score=df.test_score.values.reshape(-1,1)

interview_score=df.interview_score.values. reshape(-1,1)
salary=df.salary.values.reshape(-1,1)

reg= LinearRegression()
reg.fit(df[['experience', 'test_score', 'interview_score',]], df.salary)

```

```python
print("The model predicts a salary of", reg.predict([[2,8,8]]), "of someone with 2 yr work experiance, 9 test score, and 6 interview score")
print(reg.predict([[2,9,6]]))
print(reg.predict([[12,10,10]]))
```

```python
reg.coef_
```
